From the above Coorelation matrix it is very difficult for us to analysize what are features having high coorelation. So let us define a function to check this.

```python
def coorelation(dataset,threshold):
  col_corr = []
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[j,i])>threshold:
        col_name = corr_matrix.columns[i]
        col_corr.append(col_name)
  return col_corr
```

```python
correlation_colums_60 = coorelation(df,0.6)
len(correlation_colums_60)
```

```python
correlation_colums_50 = coorelation(df,0.5)
len(correlation_colums_50)
```

```python
correlation_colums_70 = coorelation(df,0.7)
len(correlation_colums_70)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def VIF(dataset):
  vif_df = pd.DataFrame()
  vif_df['Features'] = dataset.columns
  vif_df['VIF'] = [variance_inflation_factor(dataset.values,i) for i in range(dataset.shape[1])]
  return vif_df
```
